# üß† Support Vector Machine (SVM) - Complete Introduction

## ‚úÖ What is SVM?

Support Vector Machine (SVM) is a **supervised machine learning algorithm** used for **classification** and **regression** tasks. However, it is mostly used for **classification** problems.

At its core, SVM tries to **find the best decision boundary (also called a hyperplane)** that separates data points of different classes with the **maximum margin**.

---

## üéØ What is the Goal of SVM?

The main goal of SVM is:

> **To find a hyperplane that best separates the classes in the feature space with the maximum margin.**

In 2D, this hyperplane is just a line. In 3D or higher dimensions, it‚Äôs a plane or hypersurface.

---

## üß© Key Components of SVM

### 1. **Hyperplane**
- A decision boundary that separates different classes.
- For `n` features, the hyperplane is an `(n-1)` dimensional space.

### 2. **Margin**
- The distance between the hyperplane and the **nearest data points** from each class.
- The larger the margin, the better the generalization of the model.

### 3. **Support Vectors**
- The **data points** that lie closest to the hyperplane.
- These are the most **critical elements** of the dataset because they **define the margin**.
- If these points are removed, the margin would change.

### 4. **Kernel Function (Advanced)**
- Used when data is **not linearly separable**.
- It transforms the original input space to a higher-dimensional space to make separation easier.

---

## üìã Basic Requirements / Assumptions in SVM

| Requirement                          | Description                                                                 |
|--------------------------------------|-----------------------------------------------------------------------------|
| ‚úÖ Linearly Separable Data           | (For basic SVM) The classes should be separable by a straight line or plane. |
| ‚úÖ Proper Feature Scaling            | SVM is sensitive to feature scale, so data should be normalized or standardized. |
| ‚úÖ No Missing Values                 | SVM doesn't handle missing data natively.                                  |
| ‚úÖ Binary or Multiclass Labels       | SVM can handle both, but it uses One-vs-One or One-vs-Rest for multiclass. |

# üìè Maximal Margin Classifier (MMC)

## üîç What is the Maximal Margin Classifier?

The **Maximal Margin Classifier** is the foundational idea behind Support Vector Machines (SVMs). It is a **linear classifier** that aims to find the **hyperplane that separates two classes with the largest possible margin**.

In simple terms:

> **It chooses the boundary (hyperplane) that is farthest away from the closest data points of any class.**

These closest data points are called **support vectors**.

---

## üéØ Goal of Maximal Margin Classifier

- **To find a hyperplane that separates the classes**.
- **Maximize the margin** between the two classes.
- **Avoid misclassification** when the data is perfectly separable.

---

## üìê Margin in MMC

- Margin is the distance between the hyperplane and the **nearest data point** from either class.
- **Maximal margin** means finding the hyperplane where this distance is **as large as possible**.

Let the decision boundary (hyperplane) be:

- w ¬∑ x + b = 0
- w ¬∑ x + b = +1 (positive class)
- w ¬∑ x + b = -1 (negative class)

Distance between the margins =  
$$
\text{Margin} = \frac{2}{\|w\|}
$$

---

## üìå Conditions for MMC

Let:
- \( x_i \) be a data point
- \( y_i \in \{-1, +1\} \) be the class label

Then the constraint for correct classification with margin is:

$$
y_i (w \cdot x_i + b) \geq 1
$$

## üß† Support Vector Machines (SVM): Support Vectors and Margins

### ‚ú≥Ô∏è Support Vectors Explained

In SVM, the model aims to separate classes using a **decision boundary** (a line in 2D, a hyperplane in higher dimensions). But instead of just any line, it chooses the one that maximizes the **margin** ‚Äî the distance between the line and the closest points from each class.

Those **closest points** from each class that lie right on the margin boundaries are called **Support Vectors**.

---

### üìê Model Equation in SVM

Model (Decision Boundary):

- **Positive Support Vector boundary*
ax + by + c = +1

- **Negative Support Vector boundary**:  ax + by + c = -1


All data points from class `+1` should lie **on or beyond** the positive margin.  
All data points from class `-1` should lie **on or beyond** the negative margin.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the coefficients for the decision boundary
a = 1   # Coefficient of x
b = -1  # Coefficient of y
c = 0   # Intercept term

# Create a range of x values
x_vals = np.linspace(-5, 5, 100)

# Compute corresponding y values for decision boundary and margins
def line_eq(x, a, b, c):
    return (-a * x - c) / b

# Decision boundary: ax + by + c = 0
y_decision = line_eq(x_vals, a, b, c)

# Positive margin: ax + by + c = 1
y_pos_margin = line_eq(x_vals, a, b, c - 1)

# Negative margin: ax + by + c = -1
y_neg_margin = line_eq(x_vals, a, b, c + 1)

# Plotting
plt.figure(figsize=(10, 6))

# Decision boundary
plt.plot(x_vals, y_decision, 'k-', label='Decision Boundary: ax + by + c = 0')

# Positive margin
plt.plot(x_vals, y_pos_margin, 'r--', label='Positive Margin: ax + by + c = 1')

# Negative margin
plt.plot(x_vals, y_neg_margin, 'b--', label='Negative Margin: ax + by + c = -1')

# Fill margin area (just for visualization)
plt.fill_between(x_vals, y_pos_margin, y_neg_margin, color='gray', alpha=0.2, label='Margin')

# Adding labels and legends
plt.title("SVM: Decision Boundary with Margins")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```

### üìè Distance Between Two Parallel Lines

For two parallel lines of the form:

- \( ax + by + c_1 = 0 \)
- \( ax + by + c_2 = 0 \)

The **perpendicular distance** \( d \) between them is given by:

$$
d = \frac{|c_2 - c_1|}{\sqrt{a^2 + b^2}}
$$

---

#### üîç Explanation:

- \( a, b \) are the same for both lines, so they are parallel.
- \( c_1, c_2 \) control the shift (position) of each line.
- This formula calculates the **shortest (perpendicular) distance** between them.


### üß† SVM Optimization: Maximal Margin Classifier

We want to **maximize the margin** between two classes.

---

#### üéØ Goal:
To find the hyperplane \( ax + by + c = 0 \) such that:

- It **separates** the data perfectly (i.e., classes are linearly separable)
- The **margin** between the closest points of each class is **maximized**

---

### ‚úÖ Constraints

Let the dataset be \( \{(x_i, y_i)\}_{i=1}^n \), where:

- \( x_i \in \mathbb{R}^2 \) (feature vector)
- \( y_i \in \{-1, +1\} \) (class labels)

The **constraints** for correctly classifying each data point are:

$$
y_i (a x_{1i} + b x_{2i} + c) \geq 1
$$

This ensures all data points lie on the correct side of the margin.

---

### üìê Margin Distance

The margin (i.e., the perpendicular distance between the two support vectors) is:

$$
\text{Margin} = \frac{2}{\sqrt{a^2 + b^2}}
$$

Hence, we aim to **maximize** this value, which is equivalent to:

$$
\arg\max_{a,b,c} \frac{2}{\sqrt{a^2 + b^2}} \quad \text{subject to} \quad y_i (a x_{1i} + b x_{2i} + c) \geq 1
$$

---

### üßÆ Summary

- **Objective**: Maximize margin \( \frac{2}{\sqrt{a^2 + b^2}} \)
- **Constraint**: \( y_i (a x_{1i} + b x_{2i} + c) \geq 1 \)
- This forms a **constrained optimization problem** solved using **Quadratic Programming**


## üß† Slack Variables in SVM (Soft Margin SVM)

In real-world datasets, classes are **not always perfectly linearly separable**. This is where we introduce **Slack Variables** \( \zeta_i \) to allow some **misclassification or margin violations**.

---

### üìå Notation

Let:
- \( \zeta_i \) be the **slack variable** for the \( i^{th} \) data point
- It measures **how much** a data point **violates** the margin (or even gets misclassified)

---

### ‚úÖ Interpretation of Slack Variable \( \zeta_i \)

| Value of \( \zeta_i \) | Interpretation |
|------------------------|----------------|
| \( \zeta_i = 0 \)      | Perfectly classified and **outside the margin** (no violation) |
| \( 0 < \zeta_i < 1 \)  | Inside the margin but **correctly classified** |
| \( \zeta_i \geq 1 \)   | **Misclassified** (lies on the wrong side of the decision boundary) |

---

### üí° Geometric Constraint (Soft Margin SVM)

For each training point \( (x_i, y_i) \), the constraint becomes:

$$
y_i(w^T x_i + b) \geq 1 - \zeta_i, \quad \text{with } \zeta_i \geq 0
$$

---

### üîß Objective Function: Soft Margin SVM

We now **balance two goals**:
1. Maximize margin (minimize \( \|w\|^2 \))
2. Minimize total slack (penalize violations using \( \sum \zeta_i \))

$$
\min_{w, b, \zeta} \quad \frac{1}{2} \|w\|^2 + C \sum_{i=1}^n \zeta_i
$$

Subject to:
$$
y_i(w^T x_i + b) \geq 1 - \zeta_i, \quad \zeta_i \geq 0 \quad \forall i
$$

- \( C \) = **Regularization Parameter** (controls trade-off between margin size and slack penalty)

---

### üìâ Hinge Loss Function

$$
\text{Hinge Loss:} \quad L(y_i, f(x_i)) = \max(0, 1 - y_i f(x_i))
$$

Where \( f(x_i) = w^T x_i + b \)

---

### ‚úÖ Combined Optimization using Hinge Loss:

$$
\min_{w, b} \quad \frac{1}{2} \|w\|^2 + C \sum_{i=1}^n \max(0, 1 - y_i(w^T x_i + b))
$$

---

We now aim to **maximize the margin** while also **minimizing total slack**:

## ‚úàÔ∏è Soft Margin Plane

In Soft Margin SVM, we aim to **balance two goals**:

- **Maximize the margin** (i.e., keep the decision boundary as far as possible from both classes)
- **Minimize misclassification** using slack variables \( \zeta_i \)

---

### üéØ Optimization Objective

We aim to minimize the following expression:

$$
\text{Objective:} \quad \arg\min_{a, b, c} \left( \frac{1}{2} \sqrt{a^2 + b^2} + C \cdot \frac{1}{n} \sum_{i=1}^{n} \zeta_i \right)
$$

Where:
- \( a, b, c \) define the separating hyperplane: \( ax + by + c = 0 \)
- \( \zeta_i \) is the slack variable for the \( i^{\text{th}} \) sample
- \( C \) is the **regularization parameter** (controls the trade-off between margin and misclassification)

---

> This objective encourages a **large margin** (by minimizing \( \sqrt{a^2 + b^2} \)) while **penalizing violations** of the margin (via \( \sum \zeta_i \)).

---


## ‚öñÔ∏è Bias-Variance Tradeoff in SVM (Role of \( C \))

In **Soft Margin SVM**, the parameter \( C \) controls the **trade-off between margin size and misclassification**:

---

### üîπ High \( C \): Low Bias, High Variance

- SVM tries to **avoid misclassifying training points**
- Leads to a **smaller margin**
- **Overfits** the training data
- **Variance increases**, as small changes in data can change the boundary
- **Bias decreases**, as the model fits training data very closely

---

### üîπ Low \( C \): High Bias, Low Variance

- SVM allows **more violations** (slack variables are tolerated)
- Finds a **larger margin**
- **Underfits** the training data
- **Variance decreases**, model is more stable
- **Bias increases**, as some training points may be misclassified

---

### üìå Summary Table

| Value of \( C \) | Margin       | Misclassification Tolerance | Bias       | Variance   |
|------------------|--------------|------------------------------|------------|------------|
| High             | Small        | Low (strict)                 | Low        | High       |
| Low              | Large        | High (lenient)               | High       | Low        |

---

> **Tip**: Choose \( C \) using cross-validation to balance generalization and training accuracy.


## üß© Types of Kernels in SVM

**Kernels** in SVM allow us to **map input features into higher-dimensional spaces**, enabling classification of non-linearly separable data.

---

### 1. **Linear Kernel**
- **Equation**:  
  \( K(x, x') = x^T x' \)
- **Use Case**: When data is **linearly separable**
- **Fast and simple**

---

### 2. **Polynomial Kernel**
- **Equation**:  
  \( K(x, x') = (x^T x' + r)^d \)  
  where \( r \) is a constant and \( d \) is the degree of the polynomial
- **Use Case**: When data has curved boundaries
- **Can model more complex relationships**

---

### 3. **Radial Basis Function (RBF) / Gaussian Kernel**
- **Equation**:  
  \( K(x, x') = \exp(-\gamma \|x - x'\|^2) \)  
  where \( \gamma \) controls the spread
- **Use Case**: **Most popular** kernel for nonlinear data
- **Very flexible**, capable of modeling complex decision boundaries

---

### 4. **Sigmoid Kernel**
- **Equation**:  
  \( K(x, x') = \tanh(\alpha x^T x' + c) \)
- **Use Case**: Inspired by neural networks
- Less commonly used compared to RBF or polynomial

---

### üîç How to Choose a Kernel?

| Kernel Type   | When to Use                        |
|---------------|------------------------------------|
| Linear        | Data is linearly separable or high-dimensional |
| Polynomial    | You suspect polynomial relationships |
| RBF / Gaussian| You want to capture complex, nonlinear patterns |
| Sigmoid       | Experimental / Neural Net-like behavior |

> **Tip**: Use cross-validation to determine the best kernel for your problem!



## üéØ Why Use the Kernel Trick Instead of Manual Feature Transformation?

In SVM, we often want to map data to a **higher-dimensional space** so that it becomes **linearly separable**.

You *could* manually transform the features using nonlinear functions, but that is:

- **Computationally expensive**
- **Difficult to design correctly**
- **Infeasible** in high-dimensional or infinite-dimensional spaces

---

### üîÆ Enter the Kernel Trick

Instead of **explicitly transforming** the data, the **kernel trick** allows us to:

- Compute the **inner product in high-dimensional space**  
- **Without ever computing** the transformed coordinates!

That means:  
We work in high-dimensional space **implicitly**, saving **time and memory**.

---

### üß† Mathematically:

Let \( \phi(x) \) be the mapping to higher-dimensional space.

Instead of computing:  
$$
( \phi(x)^T \phi(x') )
$$

We use a **kernel function**:  
$$
( K(x, x') = \phi(x)^T \phi(x') )
$$

---

### ‚úÖ Benefits of Kernel Trick:

| Feature | Why It Matters |
|--------|----------------|
| **Efficiency** | No need to compute or store high-dimensional features |
| **Flexibility** | Enables use of **infinite-dimensional** spaces (like RBF kernel) |
| **Simplicity** | You just plug in a kernel function like `rbf`, `poly`, etc. |
| **Scalability** | Makes complex nonlinear decision boundaries **computationally feasible** |

---

### üß™ Example:

Instead of computing 
$$
( \phi(x) = [x_1^2, \sqrt{2}x_1x_2, x_2^2] )  
$$

You just compute:  
$$
( K(x, x') = (x^T x')^2 )
$$

Which **gives the same result**, much faster!

---

> **Conclusion**:  
The **kernel trick** makes SVMs powerful and practical for real-world nonlinear problems ‚Äî without the burden of explicitly transforming data.



## üîç Types of Kernels in SVM (Support Vector Machines)

Kernels are **functions that compute the similarity** between two data points in a higher-dimensional space ‚Äî allowing SVM to find nonlinear decision boundaries efficiently using the **kernel trick**.

---

### 1. **Linear Kernel**

**Formula:**
$$
K(x, x') = x^T x'
$$

**Use case:**
- When data is **linearly separable**
- Fastest to compute

**Pros:**
- Simple and interpretable
- Works well for text classification or linearly distributed features

---

### 2. **Polynomial Kernel**

**Formula:**
$$
K(x, x') = (x^T x' + c)^d
$$

Where:
- \( c \) = constant (controls flexibility)
- \( d \) = degree of the polynomial

**Use case:**
- When you want to model **interactions** or **nonlinear relations**

**Pros:**
- Captures feature combinations (e.g., quadratic, cubic terms)
- Good for medium complexity datasets

**Cons:**
- Can become computationally expensive for large \( d \)

---

### 3. **Radial Basis Function (RBF) / Gaussian Kernel**

**Formula:**
$$
K(x, x') = \exp\left(-\gamma \|x - x'\|^2\right)
$$

Where:
- \( \gamma \) = defines how far the influence of a training point reaches

**Use case:**
- Best for **nonlinear** data with **complex decision boundaries**

**Pros:**
- Highly flexible
- Works in **infinite-dimensional space**
- Popular default choice in most libraries (like `sklearn`)

**Cons:**
- Requires careful tuning of \( \gamma \) and \( C \)
- May overfit if not properly regularized

---

### 4. **Sigmoid Kernel**

**Formula:**
$$
K(x, x') = \tanh(\alpha x^T x' + c)
$$

Where:
- \( \alpha \), \( c \) are kernel parameters

**Use case:**
- Inspired by **neural networks** (acts like an activation function)

**Pros:**
- Models similarity like a neural network layer

**Cons:**
- Not always a valid kernel (depending on parameter values)
- Rarely used in practice compared to RBF or polynomial

---

### 5. **Custom / Precomputed Kernels**

You can define your own kernel function \( K(x, x') \) if:

- You have **domain-specific knowledge**
- You want to use **non-standard similarity measures**

**Example use cases:**
- String kernels for DNA sequences
- Tree kernels for parsing structures

---

### üìä Kernel Comparison Table

| Kernel     | Nonlinearity | Parameters             | Space Mapped To           | Use Case                         |
|------------|--------------|------------------------|----------------------------|----------------------------------|
| Linear     | No           | None                   | Original feature space     | Simple, linearly separable data |
| Polynomial | Yes          | Degree \( d \), Constant \( c \) | Finite-dimensional polynomial space | Captures interactions            |
| RBF        | Yes          | \( \gamma \)           | Infinite-dimensional       | Complex decision boundaries     |
| Sigmoid    | Yes          | \( \alpha \), \( c \)  | Related to neural nets     | Experimental, rarely used       |

---

> **Tip**: Start with **RBF Kernel**, and if performance is not good, explore **linear** or **polynomial** depending on the data's nature.



# ‚úÖ Assumptions of Support Vector Machines (SVM)

Although SVM is a powerful and flexible classifier, it does rely on a few key assumptions to work effectively:

---

## 1. **Margin Maximization**

- SVM assumes that the **best separating hyperplane** is the one that **maximizes the margin** between classes.
- This works well if the data is either linearly separable or **can be made separable** in higher dimensions using kernel tricks.

---

## 2. **Binary Classification (initially)**

- Standard SVMs are designed for **binary classification** problems.
- For multi-class tasks, strategies like **One-vs-One** or **One-vs-Rest** are used.

---

## 3. **Independently and Identically Distributed (i.i.d.) Data**

- The training data should be **independent and identically distributed**.
- Violations of this (e.g., time-series data without preprocessing) may reduce performance.

---

## 4. **Clear Margin of Separation**

- SVM performs best when there is a **clear margin** (even if soft) between classes.
- When classes are heavily overlapped, performance may degrade.

---

## 5. **Feature Scaling is Important**

- SVM assumes that all features contribute **equally** to the margin.
- Therefore, it is essential to **normalize or standardize** features (e.g., using z-score normalization).

---

## 6. **Noisy Data May Impact Performance**

- Especially in the case of a **hard-margin SVM**, noisy outliers can heavily distort the decision boundary.
- Using a **soft margin** and adjusting the regularization parameter \( C \) helps control overfitting.

---

## 7. **Kernel Trick Validity**

- When using a kernel, it must satisfy **Mercer‚Äôs Theorem** to ensure a valid transformation into a higher-dimensional space.

---

## üîç Summary Table

| Assumption                     | Description                                      |
|-------------------------------|--------------------------------------------------|
| Margin Maximization            | Optimal boundary maximizes separation margin     |
| Binary Classification          | SVM is inherently binary                         |
| i.i.d. Data                    | Data should be independently sampled             |
| Feature Scaling                | Essential to prevent bias from large features    |
| Low Noise                      | Works best on clean data or with soft margins    |
| Valid Kernel                   | Must be positive semi-definite (Mercer‚Äôs theorem)|

---


# ‚ö†Ô∏è Limitations of Support Vector Machines (SVM)

While SVMs are powerful and effective in many scenarios, they also come with several limitations:

---

## 1. **Computational Complexity**
- **Training time increases** rapidly with the size of the dataset.
- Not ideal for **very large datasets** (in terms of number of samples or features).
- Especially expensive when using non-linear kernels.

---

## 2. **Choice of Kernel**
- Performance is **highly dependent** on selecting the right kernel.
- There is **no universal rule** for choosing the best kernel or kernel parameters.
- Wrong kernel may lead to poor generalization.

---

## 3. **Difficult to Interpret**
- Compared to models like Decision Trees or Logistic Regression, SVMs are **less interpretable**.
- Not ideal when model explainability is important (e.g., in healthcare or finance).

---

## 4. **Sensitive to Noise and Outliers**
- Outliers can **significantly affect** the placement of the decision boundary.
- Although soft margin SVM helps, noisy data may still lead to **overfitting** or misclassification.

---

## 5. **Binary Classification by Default**
- Native SVM is designed for **binary classification**.
- Requires additional techniques like **One-vs-One (OvO)** or **One-vs-Rest (OvR)** for multi-class classification, which adds complexity.

---

## 6. **Feature Scaling is Mandatory**
- SVMs assume all features are on the same scale.
- Failing to standardize or normalize features can cause **biased results**.

---

## 7. **No Probabilistic Interpretation**
- SVM does **not provide probability estimates** directly.
- You need additional techniques (e.g., Platt scaling) for probabilistic output.

---

## üîç Summary Table

| Limitation                    | Description                                           |
|------------------------------|-------------------------------------------------------|
| High Computational Cost      | Slow on large datasets                               |
| Kernel Selection Sensitivity | Right kernel is crucial                              |
| Poor Interpretability        | Hard to explain decisions                            |
| Noise Sensitivity            | Outliers can affect results                          |
| Not Native Multi-Class       | Requires OvR or OvO strategies                       |
| Requires Feature Scaling     | Essential to avoid bias from unscaled features       |
| No Probabilities             | Doesn‚Äôt output probabilities without extra steps     |

---


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

# Load simple dataset
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=6)
y[y == 0] = -1  # convert labels to -1 and 1

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train linear SVM
model = SVC(kernel='linear', C=1e5)
model.fit(X_scaled, y)

# Get model parameters
w = model.coef_[0]
b = model.intercept_[0]

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap='bwr', edgecolors='k', s=70)
ax = plt.gca()

# Create a grid to evaluate model
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx = np.linspace(xlim[0], xlim[1])
yy = -(w[0] * xx + b) / w[1]

# Margin
margin = 1 / np.linalg.norm(w)
yy_down = yy - np.sqrt(1 + (w[0]/w[1])**2) * margin
yy_up = yy + np.sqrt(1 + (w[0]/w[1])**2) * margin

# Plot decision boundary and margins
plt.plot(xx, yy, 'k-', label='Decision boundary')
plt.plot(xx, yy_down, 'r--', label='-1 Margin')
plt.plot(xx, yy_up, 'g--', label='+1 Margin')

# Highlight support vectors
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
            s=150, facecolors='none', edgecolors='k', label='Support Vectors')

plt.legend()
plt.title("Linear SVM with Margin and Support Vectors")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.datasets import make_regression

# Generate synthetic regression data
X, y = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)
y = y / 100  # Scale target to make the margin effect visible

# Fit SVR models with different kernels
svr_linear = SVR(kernel='linear', C=100)
svr_poly = SVR(kernel='poly', C=100, degree=2)
svr_rbf = SVR(kernel='rbf', C=100, gamma=0.1)

# Fit the models
svr_linear.fit(X, y)
svr_poly.fit(X, y)
svr_rbf.fit(X, y)

# Predict
X_test = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_pred_linear = svr_linear.predict(X_test)
y_pred_poly = svr_poly.predict(X_test)
y_pred_rbf = svr_rbf.predict(X_test)

# Plotting
plt.figure(figsize=(16, 4))

plt.subplot(1, 3, 1)
plt.scatter(X, y, color='darkorange', label='data')
plt.plot(X_test, y_pred_linear, color='navy', lw=2, label='Linear SVR')
plt.title('Linear Kernel')
plt.legend()

plt.subplot(1, 3, 2)
plt.scatter(X, y, color='darkorange', label='data')
plt.plot(X_test, y_pred_poly, color='green', lw=2, label='Polynomial SVR')
plt.title('Polynomial Kernel')
plt.legend()

plt.subplot(1, 3, 3)
plt.scatter(X, y, color='darkorange', label='data')
plt.plot(X_test, y_pred_rbf, color='red', lw=2, label='RBF SVR')
plt.title('RBF Kernel')
plt.legend()

plt.tight_layout()
plt.show()
```

## üß† How SVM Works: Classification vs Regression

---

### üî∑ 1. SVM for Classification

**Goal**:  
To find the **optimal hyperplane** that separates the data into different classes **with the maximum margin**.

---

#### ‚úèÔ∏è Key Concepts:
- **Hyperplane**: A linear decision boundary defined as:
  $$
  w^T x + b = 0
  $$
- **Margin**: Distance between the hyperplane and the **nearest data points** (support vectors)
- **Support Vectors**: Points lying closest to the decision boundary ‚Äî they **define the margin**

---

#### ‚öôÔ∏è Objective (Soft Margin SVM):

We allow some misclassification using **slack variables** \( \zeta_i \), and solve:

$$
\min_{w, b, \zeta} \quad \frac{1}{2} \|w\|^2 + C \sum_{i=1}^{n} \zeta_i
$$

Subject to:
$$
y_i(w^T x_i + b) \geq 1 - \zeta_i, \quad \zeta_i \geq 0
$$

- \( C \): Regularization parameter controlling margin vs. misclassification
- \( \zeta_i \): Slack variable for margin violation

---

### üî∑ 2. SVM for Regression (SVR)

**Goal**:  
To find a function that **deviates from actual targets by no more than \( \epsilon \)**, while keeping the model as **flat** as possible.

---

#### ‚úèÔ∏è Key Concepts:

- **Margin of tolerance**: \( \epsilon \) defines a tube within which no penalty is given
- **Support vectors**: Points **outside the \( \epsilon \)-tube**
- **Slack variables** \( \zeta_i, \zeta_i^* \): Allow violations above and below the margin

---

#### ‚öôÔ∏è Objective (SVR):

Minimize:
$$
\frac{1}{2} \|w\|^2 + C \sum_{i=1}^{n} (\zeta_i + \zeta_i^*)
$$

Subject to:
$$
\begin{align*}
y_i - (w^T x_i + b) &\leq \epsilon + \zeta_i \\
(w^T x_i + b) - y_i &\leq \epsilon + \zeta_i^* \\
\zeta_i, \zeta_i^* &\geq 0
\end{align*}
$$

---

### ‚úÖ Summary Table

| Feature              | Classification                  | Regression (SVR)                      |
|----------------------|----------------------------------|----------------------------------------|
| Target               | Discrete class labels            | Continuous values                     |
| Margin               | Between support vectors          | \( \epsilon \)-tube around predictions |
| Slack variables      | \( \zeta_i \): margin violation  | \( \zeta_i, \zeta_i^* \): error above/below |
| Objective            | Max margin, min violations       | Min flatness, min violations           |
| Use Case             | Spam detection, image classification | Price prediction, trend forecasting |
