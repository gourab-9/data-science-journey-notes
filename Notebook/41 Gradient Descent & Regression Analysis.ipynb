### **Gradient Descent**

Gradient Descent is an optimization algorithm used to minimize a differentiable function. The core idea is to update model parameters iteratively in the direction of the negative gradient of the loss function to reach a local or global minimum.

It is not limited to **Linear Regression**; it is widely used in:
- Logistic Regression
- Deep Learning (as a core optimization backbone)
- Dimensionality reduction techniques like t-SNE

---

### **Types of Gradient Descent**

#### 1. **Batch Gradient Descent**
- **Working**: Computes the gradient using the **entire dataset** and updates parameters **once per epoch**.
- **Advantages**:
  - Stable convergence (smooth path towards minimum)
  - Accurate gradients
- **Disadvantages**:
  - Computationally expensive for large datasets
  - Requires high memory and good hardware
- **Note**: One update after seeing all training examples in a single epoch.

---

#### 2. **Stochastic Gradient Descent (SGD)**
- **Working**: Updates model parameters using the gradient of the **loss function for a single example**.
- **Advantages**:
  - Faster convergence for large datasets
  - Requires less memory and hardware
- **Disadvantages**:
  - Noisy updates may cause fluctuation around the minimum
  - Non-deterministic: Different runs may give different results
- **Note**: Frequent updates lead to faster but less stable convergence.

---

#### 3. **Mini-Batch Gradient Descent**
- **Working**: Splits the training data into **small batches** and performs updates on each mini-batch.
- **Advantages**:
  - Balanced trade-off between speed and stability
  - More efficient than batch, less noisy than SGD
- **Example**: If you have 1000 data points and a batch size of 100, you get 10 mini-batches, resulting in 10 updates per epoch.
- **Note**: Reduces variance and offers a more stable path toward convergence.

---

### **Summary**

| Type                     | Data Used per Update | Speed     | Stability | Hardware Requirement |
|--------------------------|----------------------|-----------|-----------|-----------------------|
| Batch Gradient Descent   | Full Dataset         | Slow      | High      | High                  |
| Stochastic Gradient Descent | 1 Sample          | Fast      | Low       | Low                   |
| Mini-Batch Gradient Descent | Batch of Samples  | Moderate  | Moderate  | Moderate              |

## 📘 What is F-Statistic in Linear Regression?

The **F-statistic** is a value that helps us determine whether the **overall regression model** is **statistically significant** — i.e., whether **at least one predictor variable (X)** has a linear relationship with the **dependent variable (Y)**.

---

## ❓ Why is F-Statistic Needed?

When you have **multiple predictors**, it's not enough to look at individual p-values of coefficients — you also need to assess whether the model **as a whole** is useful.

The F-test answers:
> Is there **at least one** independent variable that explains a significant portion of the variation in the dependent variable?

---

## 🧠 Intuition

- A regression model that fits the data well should reduce error.
- The F-statistic compares the model with **all features** (predictors) against a **baseline model** that uses **no features** (only intercept).
- If adding features significantly reduces the error, the F-statistic will be **large**.

---

## 📐 Formula

Let:
- SSR = Sum of Squares due to Regression
- SSE = Sum of Squares due to Error (residual)
- n = number of observations
- k = number of predictors

$$
F = \frac{(SSR / k)}{(SSE / (n - k - 1))}
$$

Or:

$$
F = \frac{\text{Explained variance per degree of freedom}}{\text{Unexplained variance per degree of freedom}}
$$

---

## ✅ Interpretation

- **High F-statistic** → Model is statistically significant
- **Low p-value** for F-statistic → Reject the null hypothesis that "all coefficients are zero"
- **F close to 1** → Model is not significantly better than baseline

---

## 🧪 Hypotheses of F-Test

- **Null Hypothesis (H₀)**: All coefficients (except intercept) = 0 → Model explains nothing
- **Alternative Hypothesis (H₁)**: At least one coefficient ≠ 0 → Model explains something

If **p-value < 0.05**, reject H₀ → Model has predictive power.

---

## 🛠 Example in Words

Suppose you build a model to predict house price using:
- Size
- Location score
- Age of house

The **F-statistic** will test whether this combination **collectively** helps in predicting house price, rather than just random noise.

---

## 📌 Summary

| Point | Description |
|-------|-------------|
| What | Measures overall significance of the regression model |
| Why  | To ensure the model is better than a naive mean-based prediction |
| When | Especially important when there are multiple predictors |
| Output | Found in OLS summary under `F-statistic` and `Prob(F-statistic)` |

---

🗒️ **Note:** Even if some individual predictors are not significant (high p-value), the overall model might still be useful, as revealed by a significant F-test.


## 📊 TSS, RSS, and ESS in Linear Regression

These three quantities help us evaluate how well a regression model fits the data:

---

### 🔷 1. Total Sum of Squares (TSS)

- **Definition**: The total variation in the observed data.
- **Represents**: How much the actual values (`Y_actual`) vary from their mean.
- **Formula**:
  
$$
TSS = \sum_{i=1}^{n} (y_i - \bar{y})^2
$$

Where:
- \( y_i \) = actual value
- \( \bar{y} \) = mean of actual values

---

### 🔷 2. Explained Sum of Squares (ESS) or Regression Sum of Squares

- **Definition**: The portion of TSS explained by the regression model.
- **Represents**: How much the predicted values differ from the mean of actual values.
- **Formula**:

$$
ESS = \sum_{i=1}^{n} (\hat{y}_i - \bar{y})^2
$$

Where:
- \( \hat{y}_i \) = predicted value by the model
- \( \bar{y} \) = mean of actual values

---

### 🔷 3. Residual Sum of Squares (RSS)

- **Definition**: The portion of TSS not explained by the model.
- **Represents**: The errors — how far the predicted values are from the actual values.
- **Formula**:

$$
RSS = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

Where:
- \( y_i \) = actual value
- \( \hat{y}_i \) = predicted value

---

### 🔁 Relationship between TSS, ESS, and RSS

$$
TSS = ESS + RSS
$$

This shows how the total variance (TSS) is divided into:
- **Explained** variance (ESS) — captured by the model
- **Unexplained** variance (RSS) — due to error

---

### 📌 Use in \( R^2 \) (Coefficient of Determination)

$$
R^2 = \frac{ESS}{TSS} = 1 - \frac{RSS}{TSS}
$$

- \( R^2 \in [0, 1] \)
- Tells how much variance in the target variable is explained by the model.

---

### 💡 Intuition

- **TSS**: Total variability in the data (baseline error without a model)
- **ESS**: What the model can explain
- **RSS**: What the model fails to explain

Lower RSS and higher ESS → better model.



````python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 10)
y_actual = 2 * x + 1 + np.random.normal(0, 3, size=x.shape)  # actual values with noise
y_predicted = 2 * x + 1  # perfect prediction without noise
y_mean = np.mean(y_actual)

# Plot setup
plt.figure(figsize=(10, 6))
plt.plot(x, y_predicted, color='blue', label='Predicted Line (ŷ)', linewidth=2)
plt.axhline(y_mean, color='black', linestyle='--', label='Mean of Actual (ȳ)')

# Plot actual points
plt.scatter(x, y_actual, color='red', label='Actual Data Points')

# Draw lines for ESS and RSS
for i in range(len(x)):
    # ESS: vertical line from mean to predicted
    plt.plot([x[i], x[i]], [y_mean, y_predicted[i]], color='green', linestyle='--')
    # RSS: vertical line from actual to predicted
    plt.plot([x[i], x[i]], [y_actual[i], y_predicted[i]], color='orange', linestyle=':')

# Labels and legend
plt.title('Visualizing TSS, ESS, and RSS')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

## 🎯 Visual Intuition of TSS, ESS, and RSS

In a regression context, we can decompose the total variation in the target variable into two parts:

---

### 📌 Definitions

- **TSS (Total Sum of Squares):**  
  The total variation in actual values \( y_i \) around their mean \( \bar{y} \).  
  $$
  \text{TSS} = \sum_{i=1}^{n} (y_i - \bar{y})^2
  $$

- **ESS (Explained Sum of Squares):**  
  The variation explained by the model — difference between predicted values \( \hat{y}_i \) and mean \( \bar{y} \).  
  $$
  \text{ESS} = \sum_{i=1}^{n} (\hat{y}_i - \bar{y})^2
  $$

- **RSS (Residual Sum of Squares):**  
  The variation not explained by the model — difference between actual and predicted values.  
  $$
  \text{RSS} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
  $$

---

### 🔗 Relationship

$$
\text{TSS} = \text{ESS} + \text{RSS}
$$

This decomposition helps understand the quality of a regression model.

---

### 🧠 Visual Breakdown

- **TSS**: Total variability in actual values around their mean.  
  → Represented as **black dashed lines** from each \( y_i \) to \( \bar{y} \).

- **ESS**: Variability **explained** by the model.  
  → Represented as **green dashed lines** from \( \bar{y} \) to \( \hat{y}_i \).

- **RSS**: Variability **not explained** by the model (residuals).  
  → Represented as **orange dotted lines** from \( \hat{y}_i \) to \( y_i \).

---

📈 A diagram showing the actual points, mean line, predicted line, and these deviations can greatly improve interpretation.



## 📊 R-Squared (Coefficient of Determination)

### 🔢 Formula

R-Squared measures how much of the total variance in the target variable \( y \) is explained by the regression model.

There are three equivalent formulas:

$$
R^2 = \frac{ESS}{TSS} = \frac{TSS-RSS}{TSS}= 1 - \frac{RSS}{TSS}
$$

Where:
- **TSS** = Total Sum of Squares = \( \sum (y_i - \bar{y})^2 \)
- **ESS** = Explained Sum of Squares = \( \sum (\hat{y}_i - \bar{y})^2 \)
- **RSS** = Residual Sum of Squares = \( \sum (y_i - \hat{y}_i)^2 \)

---

### 🎯 Interpretation

- \( R^2 = 1 \): The model explains **100%** of the variance in the data (perfect fit).
- \( R^2 = 0 \): The model does **not** explain any variance better than the mean.
- \( R^2 < 0 \): The model performs **worse** than a horizontal mean line (poor fit).

$$
R^2 = \text{Proportion of the total variance in } y \text{ that is explained by } \hat{y}
$$

---

### ✅ Advantages

- Intuitive interpretation: What **% of variability** in the output is captured by the model.
- Easy to compare two models using their \( R^2 \) values (e.g., 0.81 > 0.69).

---

### ⚠️ Limitations

- **Misleading with many features**: 
  Adding more features (even irrelevant ones) **never decreases** \( R^2 \). It either increases or remains the same.
  
  That means a model can look artificially good just by including extra columns — even if they don’t actually help.

- This is why **Adjusted R-Squared** is preferred in such cases — it penalizes the addition of non-informative variables.

---

### ℹ️ Use Cases

- Great for simple linear models with a few meaningful predictors.
- Not reliable for feature-rich models — use Adjusted \( R^2 \) or other metrics like AIC/BIC or cross-validated RMSE.



## 📐 Adjusted R-Squared

### 🧮 Formula

$$
\text{Adjusted } R^2 = 1 - \left( \frac{(1 - R^2)(n - 1)}{n - k - 1} \right)
$$

Where:
- \( R^2 \) = R-squared of the model
- \( n \) = Number of observations (rows of data)
- \( k \) = Number of predictor variables (independent features)

---

### ✅ Why Adjusted R²?

While \( R^2 \) always increases (or stays the same) when you add more features, **Adjusted \( R^2 \)** corrects for that by adding a penalty for irrelevant (non-informative) features.

It **only increases** if the new feature improves the model **more than expected by chance**.

---

### 🎯 Interpretation

- Adjusted \( R^2 \) < \( R^2 \) if added features are irrelevant.
- Adjusted \( R^2 \) ≈ \( R^2 \) if the model is well-balanced.
- If you add a **useful feature**, both numerator and denominator increase proportionally — Adjusted \( R^2 \) stays steady or increases.
- If you add an **irrelevant feature**, numerator stays the same but denominator increases — Adjusted \( R^2 \) drops.

---

### ⚠️ Why It Matters?

- Helps detect **overfitting**: Adding noisy or irrelevant columns increases model complexity without real gains.
- Prevents misleadingly high \( R^2 \) from many unimportant predictors.
- Especially useful in **multiple linear regression** with many features.

---

### 📌 Extra Notes

- Even if your **sample data looks linear** (and hence high adjusted \( R^2 \)), it may not generalize well to the population if the **true underlying pattern is nonlinear**.
- That’s why we often combine this with **F-statistics** and **residual analysis**.

---

### 🧠 Remember

Adjusted \( R^2 \) is better than plain \( R^2 \) when:
- You have **many predictors**
- You are comparing **models with different numbers of features**
- You want to **penalize irrelevant variables**



## 📐 Adjusted R-Squared and Related Concepts

### 🧮 Formula

$$
\text{Adjusted } R^2 = 1 - \left( \frac{(1 - R^2)(n - 1)}{n - k - 1} \right)
$$

Where:
- \( R^2 \) = Coefficient of determination
- \( n \) = Number of observations (rows)
- \( k \) = Number of predictor variables

---

### ✅ Why Adjusted \( R^2 \)?

- Plain \( R^2 \) always increases with more features, even irrelevant ones.
- Adjusted \( R^2 \) adds a **penalty** for unnecessary predictors.
- It reflects how well the model generalizes, not just fits the sample.

---

### 🎯 Interpretation

- **Increases** only if the new feature adds genuine explanatory power.
- **Decreases** if the new feature adds noise or overfits.
- **Helps choose a simpler model** that explains the data without overfitting.

---

### ⚠️ Key Concepts to Understand It

#### 1. **Noise / Irreducible Error**
- Every real-world dataset has randomness or noise that **no model can explain**.
- This is called **irreducible error** — it cannot be removed, only minimized.

#### 2. **Standard Error**
- The **standard deviation** of the estimate (like a slope or intercept).
- A small standard error means the estimate is reliable.
- Useful for constructing **confidence intervals**.

#### 3. **F-Statistic**
- Even if Adjusted \( R^2 \) is good, we ask:
  > Is the model significantly better than just predicting the mean?

- F-statistic compares:
  - **Model vs. Baseline (mean model)**
  - It checks if **the relationship observed in the sample** is likely to hold for the **population**.
- Why needed?
  - Because **Adjusted \( R^2 \) can be high** in a sample, even if the **true pattern in the population is non-linear or weak**.

---

### 🧠 Summary

| Concept              | Purpose                                                                 |
|----------------------|-------------------------------------------------------------------------|
| Adjusted \( R^2 \)   | Penalizes unnecessary predictors to avoid overfitting                   |
| Irreducible Error    | Random variation in data that no model can explain                      |
| Standard Error       | Precision of estimated coefficients                                     |
| F-Statistic          | Tests if the model’s explanatory power is statistically significant     |

> **Use Adjusted \( R^2 \)** for fair model comparison,  
> **Standard Error** for reliability, and  
> **F-statistic** for statistical significance.


## 📊 F-Statistic in Regression

The **F-statistic** tests whether your overall regression model is a good fit for the data.

---

### 📌 What Is It?

The F-statistic answers:
> Is there a significant linear relationship between the **independent variables** and the **dependent variable**?

In other words:
> Is your regression model better than a model that just uses the **mean of the dependent variable**?

---

### 📐 F-Statistic Formula

$$
F = \frac{\text{Explained Variance per degree of freedom}}{\text{Unexplained Variance per degree of freedom}} = \frac{ESS/k}{RSS/(n - k - 1)}
$$

Where:
- \( ESS \): Explained Sum of Squares
- \( RSS \): Residual Sum of Squares
- \( n \): Total observations
- \( k \): Number of predictors

---

### 🧠 Intuition

| Component               | Interpretation                                                  |
|-------------------------|------------------------------------------------------------------|
| **Numerator (ESS/k)**   | Measures how much variation is explained by the model            |
| **Denominator (RSS/dof)**| Measures how much variation is still unexplained (errors)        |

If your model **explains a lot** compared to what it doesn't, F will be **high** → indicating statistical significance.

---

### ✅ When Is It Significant?

- You compare the calculated **F-statistic** with a critical value from the **F-distribution**.
- Alternatively, you check the **p-value** associated with the F-stat in regression summary.
- If **p < 0.05**, the regression model is **statistically significant** overall.

---

### 📈 Example in OLS Summary

```text
                            OLS Regression Results                            
==============================================================================
Dep. Variable:                      y   R-squared:                       0.85
Model:                            OLS   Adj. R-squared:                  0.84
Method:                 Least Squares   F-statistic:                     55.23
Date:                Tue, 30 Apr 2024   Prob (F-statistic):           1.20e-10
==============================================================================


### 🔍 F-Statistic Interpretation

**F-statistic = 55.23** → High value suggests a strong model  
**Prob(F-statistic) = 1.20e-10** → Very small p-value → The model is statistically significant

---

### 🚦 Important Notes

- A **high F-statistic with a low p-value** means the model likely fits the data well.
- If the **F-statistic is low**, your model may not be better than just using the **mean of Y**.
- F-statistic remains important **even when Adjusted R² is good**, because:
  - Your **sample may appear linear**, but the **population may not** be.
  - The **F-test helps statistically verify model strength** beyond sample data.

---

### 🧪 Summary Table

| Term             | Meaning                                                           |
|------------------|-------------------------------------------------------------------|
| **F-statistic**  | Ratio of explained to unexplained variance (normalized)          |
| **p-value**      | Significance of the overall regression model                      |
| **High F + Low p**| Your regression model explains a significant portion of variance |
