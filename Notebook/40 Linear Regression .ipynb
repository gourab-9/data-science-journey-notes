# **Linear Regression**

---

## **What is Linear Regression?**

- **Linear Regression** is a **supervised machine learning algorithm** used for predicting a continuous output based on one or more input features.
- It models the relationship between independent variables (inputs) and a dependent variable (output) by fitting a linear equation to the observed data.

---

## **Types of Linear Regression**

1. **Simple Linear Regression**:
   - Involves **one input feature** (independent variable) and **one output** (dependent variable).
   - Example: Predicting salary based on years of experience.

2. **Multiple Linear Regression**:
   - Involves **two or more input features** and **one output**.
   - Example: Predicting house price based on size, location, and number of rooms.

3. **Polynomial Linear Regression**:
   - A type of linear regression where the relationship between the independent variable and dependent variable is modeled as an **n-th degree polynomial**.
   - Although it fits a curve, it is still considered linear regression because the model is **linear in parameters**.

---

## **Linear Regression Equation**

The general equation of a simple linear regression model is:

$$
y = mx + b
$$

Where:
- \( y \) = output (predicted value),
- \( x \) = input feature,
- \( m \) = slope (coefficient),
- \( b \) = intercept (bias).

In multiple linear regression (with multiple inputs), the equation extends to:

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n
$$

---

## **How to Find the Best Parameters (m and b)?**

There are two main methods:

1. **Closed-form Solution** (Direct calculation):
   - Also called **Ordinary Least Squares (OLS)**.
   - Formula to find coefficients:

$$
\boldsymbol{\beta} = (X^T X)^{-1} X^T y
$$

Where:
- \( X \) = matrix of input features,
- \( y \) = vector of outputs,
- \( \boldsymbol{\beta} \) = vector of coefficients (parameters).

2. **Non-closed-form Solution** (Iterative approach):
   - **Gradient Descent Algorithm**.
   - Starts with random parameter values and iteratively updates them to minimize the error (cost function).

---

## **Key Points about OLS and Gradient Descent**

| Method              | Description |
|:--------------------|:------------|
| **OLS (Ordinary Least Squares)** | Direct mathematical formula to compute the best-fit line. Used when data is small to medium-sized. |
| **Gradient Descent** | Iterative optimization algorithm. Useful when data is very large and matrix inversion in OLS becomes expensive. |

- **Scikit-learn**'s `LinearRegression` model uses **OLS** internally.
- `SGDRegressor` in Scikit-learn uses **Stochastic Gradient Descent** to fit the model.

---

## **Quick Note on Terms**

- \( \bar{y} \) = Mean of actual output values.
- \( y_i \) = Actual value for the \( i \)-th data point.
- \( \hat{y}_i \) = Predicted value for the \( i \)-th data point.

# **Regression Metrics: Formulas, Advantages, Disadvantages, and Flaws**

---

## **1. Mean Absolute Error (MAE)**

### **Formula**:
$$
MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$
Where:
- \( y_i \) is the actual value
- \( \hat{y}_i \) is the predicted value
- \( n \) is the number of data points

### **Advantages**:
- **Easy to interpret**: Provides the average magnitude of errors in the same unit as the output.
- **Robust to outliers**: MAE treats all errors equally and doesn't disproportionately penalize outliers.

### **Disadvantages**:
- **Doesn’t penalize large errors**: MAE doesn’t distinguish between small and large errors; both are treated equally.
- **Non-differentiable at zero**: The absolute value function isn’t differentiable at zero, making gradient-based optimization difficult.

### **Flaws**:
- **Lack of sensitivity to outliers**: It may be unsuitable when large errors should be penalized more (e.g., when outliers are important to your analysis).

---

## **2. Mean Squared Error (MSE)**

### **Formula**:
$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$
Where:
- \( y_i \) is the actual value
- \( \hat{y}_i \) is the predicted value
- \( n \) is the number of data points

### **Advantages**:
- **Sensitive to large errors**: MSE penalizes large deviations more heavily, which is useful when large errors are more critical.
- **Differentiable**: MSE is continuously differentiable, making it suitable for optimization algorithms like gradient descent.

### **Disadvantages**:
- **Sensitive to outliers**: Large errors will dominate the calculation, making MSE very sensitive to outliers.
- **Harder to interpret**: The units of MSE are squared, which can make the results harder to interpret compared to the original data.

### **Flaws**:
- **Overemphasis on outliers**: MSE can lead to overfitting, especially when the data contains outliers, as it heavily penalizes larger errors.

---

## **3. Root Mean Squared Error (RMSE)**

### **Formula**:
$$
RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$$
Where:
- \( y_i \) is the actual value
- \( \hat{y}_i \) is the predicted value
- \( n \) is the number of data points

### **Advantages**:
- **Interpretability**: RMSE provides error values in the same units as the output variable, making it easier to interpret.
- **Sensitive to large errors**: Like MSE, RMSE penalizes larger errors, which is important in certain use cases.

### **Disadvantages**:
- **Still sensitive to outliers**: RMSE shares the same issue as MSE, as it is affected by large errors.
- **Overemphasis on large deviations**: RMSE may lead to models that overemphasize the importance of larger errors.

### **Flaws**:
- **Increased sensitivity to outliers**: Similar to MSE, it overemphasizes large errors, which might lead to biased results when data contains extreme values.

---

## **4. R-Squared (R²)**

### **Formula**:
$$
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2_{\text{reg}}}{\sum_{i=1}^{n} (y_i - \bar{y})^2_{\text{mean}}}
$$

Where:
- \( y_i \) is the **actual value** (the observed value of the dependent variable),
- \( \hat{y}_i \) is the **predicted value** (the estimated value of the dependent variable from the model),
- \( \bar{y} \) is the **mean of the actual values** (the average of all observed values of \( y \)).
s

### **Advantages**:
- **Good indicator of model fit**: R² indicates how well the independent variables explain the variance in the dependent variable.
- **Widely used**: It’s one of the most commonly used metrics to evaluate regression models.

### **Disadvantages**:
- **Doesn't account for overfitting**: R² can be misleading if the model is overfitting, as it will always increase with more predictors, regardless of their relevance.
- **Doesn’t apply to non-linear models**: R² doesn’t provide an accurate measure of fit for non-linear regression models.

### **Flaws**:
- **Misleading with complex models**: In more complex models (like polynomial regression), R² may give a falsely high value, even if the model is not generalizing well.

---

## **5. Adjusted R-Squared (Adjusted R²)**

### **Formula**:
$$
\text{Adjusted } R^2 = 1 - \left( 1 - R^2 \right) \frac{n-1}{n-p-1}
$$
Where:
- \( n \) is the number of observations
- \( p \) is the number of predictors

### **Advantages**:
- **Corrects for overfitting**: Adjusted R² penalizes the addition of irrelevant predictors and prevents overfitting by adjusting for the number of predictors in the model.
- **More accurate for multiple regression**: It provides a more accurate measure of fit when dealing with multiple regression models.

### **Disadvantages**:
- **Still susceptible to poor model fit**: Although adjusted R² corrects for the number of predictors, it can still mislead if the model isn’t appropriate for the data.
- **Not informative on its own**: The adjusted R² value needs to be interpreted alongside other metrics to understand model performance fully.

### **Flaws**:
- **Misleading for complex models**: Like R², adjusted R² can still be misleading for non-linear regression or overfitted models.

---

## **6. Mean Absolute Percentage Error (MAPE)**

### **Formula**:
$$
MAPE = \frac{1}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right| \times 100
$$
Where:
- \( y_i \) is the actual value
- \( \hat{y}_i \) is the predicted value
- \( n \) is the number of data points

### **Advantages**:
- **Easy to interpret**: MAPE expresses errors as percentages, making it easy to understand and communicate.
- **Comparative**: Since the error is relative, it allows comparisons between models and datasets of different scales.

### **Disadvantages**:
- **Not defined for zero**: If the actual value \( y_i \) is zero, MAPE becomes undefined.
- **Sensitive to outliers**: Large errors with small actual values can result in very large percentages, making MAPE sensitive to outliers.

### **Flaws**:
- **Problem with zero values**: MAPE can be invalid if the actual values contain zeros, which is common in some real-world datasets.

---

## **7. Huber Loss**

### **Formula**:
$$
L_{\delta}(y, \hat{y}) =
\begin{cases}
\frac{1}{2} (y - \hat{y})^2 & \text{for } |y - \hat{y}| \leq \delta \\
\delta \left( |y - \hat{y}| - \frac{1}{2} \delta \right) & \text{otherwise}
\end{cases}
$$
Where:
- \( \delta \) is a threshold parameter

### **Advantages**:
- **Combination of MSE and MAE**: Huber loss combines the advantages of both MAE and MSE, making it less sensitive to outliers than MSE while still penalizing large errors.
- **Robust**: It is robust to outliers because it switches from quadratic to linear loss beyond a threshold.

### **Disadvantages**:
- **Requires tuning**: The threshold \( \delta \) needs to be selected, which might require cross-validation.
- **Not always better than MSE**: For certain problems, MSE may still outperform Huber loss.

### **Flaws**:
- **Sensitive to \( \delta \) selection**: The choice of \( \delta \) can significantly affect performance, and selecting an appropriate value requires experimentation.

---

### **Conclusion:**
Each regression metric has its strengths and weaknesses, and their suitability depends on the problem at hand. MAE is useful for simpler problems where large errors should not dominate, while MSE and RMSE are better for situations where large errors are crucial. R² and Adjusted R² are valuable for understanding model fit but must be interpreted carefully, especially when overfitting is a concern.

# Multiple Linear Regression 

---

## 1. Definition

**Multiple Linear Regression (MLR)** models the relationship between a dependent variable and multiple independent variables, assuming a linear relationship between them.

---

## 2. Model Equation

The equation is:

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \beta_3 x_3 + \cdots + \beta_n x_n
$$

Or, in **matrix form**:

$$
\mathbf{X} =
\begin{bmatrix}
1 & x_{11} & x_{12} & x_{13} & \cdots & x_{1n} \\
1 & x_{21} & x_{22} & x_{23} & \cdots & x_{2n} \\
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots \\
1 & x_{m1} & x_{m2} & x_{m3} & \cdots & x_{mn} \\
\end{bmatrix}
$$

$$
\boldsymbol{\beta} =
\begin{bmatrix}
\beta_0 \\
\beta_1 \\
\beta_2 \\
\beta_3 \\
\vdots \\
\beta_n \\
\end{bmatrix}
$$

Thus:

$$
\hat{y} = X\boldsymbol{\beta}
$$

Where:

$$
X \in \mathbb{R}^{m \times (n+1)}
$$

$$
\boldsymbol{\beta} \in \mathbb{R}^{(n+1) \times 1}
$$

$$
\hat{y} \in \mathbb{R}^{m \times 1}
$$

---

## 3. Geometric Intuition

- **Each data point** \((x_1, x_2, \dots, x_n, y)\) lies in **(n+1)-dimensional space**.
- We are trying to fit a **hyperplane** that best approximates the data points.
- In **2D**, it is a line;  
  in **3D**, it is a plane;  
  in **higher dimensions**, it is a hyperplane.

**Imagine:**
- Data points scattered in space.
- The hyperplane tries to "balance" itself to **minimize vertical distances** (errors) from the points.

---

## 4. Loss Function (Residual Sum of Squares)

The **residual** for each point is:

$$
e_i = y_i - \hat{y}_i
$$

The **total error** (Loss \( \epsilon \)) is:

$$
\epsilon = \sum_{i=1}^{m} (y_i - \hat{y}_i)^2
$$

In **matrix form**:

$$
\epsilon = (Y - \hat{Y})^T (Y - \hat{Y})
$$

Expanding:

$$
\epsilon = Y^T Y - 2Y^T \hat{Y} + \hat{Y}^T \hat{Y}
$$

---

## 5. Optimization Idea

- We want to **find the best** \( \boldsymbol{\beta} \) **that minimizes the loss** \( \epsilon \).
- Setting the derivative \( \frac{\partial \epsilon}{\partial \boldsymbol{\beta}} = 0 \), we get the **Normal Equation**.

Thus, the solution for \( \boldsymbol{\beta} \) is:

$$
\boldsymbol{\beta} = (X^T X)^{-1} X^T Y
$$

**Important:**  
- \( X^T X \) must be **invertible**.
- Otherwise, we use **pseudoinverse** or **regularization techniques** (like Ridge Regression).

---

### ⚡ Computational Note

While solving the OLS (Ordinary Least Squares) solution, you need to compute the inverse of the matrix \( (X^T X) \).

- Finding the inverse of a matrix has a **time complexity of** \( \mathcal{O}(n^3) \), where \( n \) is the number of features.
- Thus, for large datasets, directly computing the inverse becomes **computationally expensive and inefficient**.
- In practice, modern libraries (like NumPy, scikit-learn) avoid direct inversion and use **faster numerical methods**, such as:
  - **QR decomposition**
  - **Singular Value Decomposition (SVD)**
  - **Gradient-based optimization**

✅ Thus, although the theory uses matrix inversion, practical implementations prefer **stable and efficient alternatives**.

---

## 6. Summary

- **Goal:** Predict \( y \) using a linear combination of features \( x_1, x_2, \dots, x_n \).
- **Model:** \( \hat{y} = X\boldsymbol{\beta} \)
- **Loss Function:** Sum of squared residuals.
- **Optimization:** Solve for \( \boldsymbol{\beta} \) by minimizing the loss using the Normal Equation.

---

# Visualization Reminder

- **Data points** are scattered in (n+1) dimensions.
- **The fitted hyperplane** tries to minimize the vertical distances (errors).
- **Best-fit plane** minimizes the overall **Sum of Squared Errors (SSE)**.

# Important Notes

- **Assumptions:**  
  Linearity, Independence of errors, Homoscedasticity (constant variance), No multicollinearity, Normality of errors.

- **Use Cases:**  
  Predicting prices, sales forecasting, risk modeling, medical diagnosis, etc.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate some synthetic data
np.random.seed(42)
x1 = np.random.rand(50) * 10
x2 = np.random.rand(50) * 10
noise = np.random.randn(50) * 2

# Define the true relationship (for synthetic data)
y = 3 + 1.5 * x1 + 2 * x2 + noise

# Fit a plane using linear regression coefficients (for simplicity using np.linalg.lstsq)
X = np.vstack((np.ones_like(x1), x1, x2)).T
beta, _, _, _ = np.linalg.lstsq(X, y, rcond=None)

# Create a meshgrid for the plane
x1_surf, x2_surf = np.meshgrid(np.linspace(x1.min(), x1.max(), 10),
                               np.linspace(x2.min(), x2.max(), 10))
y_surf = beta[0] + beta[1]*x1_surf + beta[2]*x2_surf

# Plotting
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the actual points
ax.scatter(x1, x2, y, color='blue', label='Data Points')

# Plot the fitted plane
ax.plot_surface(x1_surf, x2_surf, y_surf, color='orange', alpha=0.5, label='Fitted Plane')

# Labels
ax.set_xlabel('X1 (Feature 1)')
ax.set_ylabel('X2 (Feature 2)')
ax.set_zlabel('Y (Target)')
ax.set_title('Multiple Linear Regression: 3D Plane Fitting')

# Legend
# (Workaround for 3D plot legend)
scatter_proxy = plt.Line2D([0],[0], linestyle="none", c='blue', marker = 'o')
plane_proxy = plt.Rectangle((0,0),1,1,fc="orange", alpha=0.5)
ax.legend([scatter_proxy, plane_proxy], ['Data Points', 'Fitted Plane'])

plt.show()
```

# **Optimization in Machine Learning**

---

## **1. What is a Function?**

A **function** is a relationship where some input(s) are mapped to an output using a specific rule or logic.

Example:
- Input → Function → Output  
- \( x \rightarrow f(x) = 2x + 3 \rightarrow y \)

---

## **2. What is a Multivariate Function?**

A **multivariate function** takes more than one input variable.

Example:  
\( f(x, y) = 2x + 3y + 1 \)

Here, the output depends on both \( x \) and \( y \).

---

## **3. What is a Parameter in a Function?**

A **parameter** is a variable within the function that determines the behavior of the output.

Example:  
In \( f(x) = mx + b \), \( m \) and \( b \) are parameters.

---

## **4. Parametric vs Non-Parametric ML Models**

### **Parametric vs Non-Parametric Models**

| Feature                      | Parametric Models                                           | Non-Parametric Models                                          |
|-----------------------------|-------------------------------------------------------------|----------------------------------------------------------------|
| **Assumptions**             | Assumes a fixed functional form or data distribution        | Makes minimal or no assumptions about data distribution        |
| **Number of Parameters**    | Fixed number, independent of dataset size                   | Grows with dataset size                                        |
| **Flexibility**             | Less flexible, may underfit complex data                    | More flexible, can model complex patterns                      |
| **Training Speed**          | Generally faster to train                                   | Can be slower with large data                                  |
| **Interpretability**        | Easier to interpret                                         | Often harder to interpret                                      |
| **Examples**                | Linear Regression, Logistic Regression, Naive Bayes,        | Decision Tree, K-NN, Random Forest,                            |
|                             | SVM (linear kernel), Perceptron                             | SVM (RBF kernel), Neural Networks                              |

---

## **5. Loss Function**

A **Loss Function** measures the difference between predicted and actual values.

**Example (Mean Squared Error):**

$$
L = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

Where:
- \( y_i \) is the actual value
- \( \hat{y}_i \) is the predicted value
- \( n \) is the number of samples

---

## **6. Optimization**

Optimization is the process of **minimizing the loss function** to improve model accuracy.

We aim to find the best parameters \( \theta \) such that:

$$
\theta = \arg\min_{\theta} \text{Loss}(\theta)
$$

---

## **7. Convex vs Non-Convex Loss Functions**

### **Convex Function**
- Has a single global minimum.
- Easy to optimize.
- Any line segment between two points lies **above the curve**.

**Example:**  
$$
f(x) = x^2
$$

Graph: (U-shaped curve)

### **Non-Convex Function**
- Can have multiple local minima and maxima.
- Harder to optimize.
- Line segment may go **below the curve**.

**Example:**  
$$
f(x) = \sin(x) + x^2
$$

Graph: (Wavy curve with many dips and peaks)

# 📘 Differentiation for Machine Learning and Data Science

---

## 🔍 What is Differentiation?

Differentiation is a fundamental concept in calculus that deals with how a function changes as its input changes. In simple terms, it gives the **rate of change** or the **slope** of a function at any point.

---

## ❓ Why is Differentiation Important in ML?

Differentiation is used to:

- **Optimize models** (e.g., minimize loss functions).
- **Update parameters** in training via gradient descent.
- **Understand curvature** (concavity/convexity) of functions.
- **Perform backpropagation** in neural networks.

---

## ⚙️ How Does Differentiation Work?

We compute the **derivative** of a function, which tells us how the output changes for a small change in input.

- If \( y = f(x) \), then the derivative is:

  $$
  \frac{dy}{dx} = f'(x)
  $$

- It represents the slope of the tangent to the curve at point \( x \).

---

## 📐 Geometric Interpretation

- The derivative at a point is the **slope of the tangent line** to the curve at that point.
- A positive derivative → function increasing.
- A negative derivative → function decreasing.
- Zero derivative → local maxima/minima (flat tangent).

---

## 🔁 Higher-Order Derivatives

- **First Derivative** \( \frac{dy}{dx} \): Rate of change.
- **Second Derivative** \( \frac{d^2y}{dx^2} \): Curvature or acceleration.
  - If \( \frac{d^2y}{dx^2} > 0 \), it's a convex function (bowl-shaped).
  - If \( \frac{d^2y}{dx^2} < 0 \), it's concave (hill-shaped).

---

## 🧮 Differentiation Rules

### 1. **Power Rule**

If \( f(x) = x^n \), then:

$$
\frac{d}{dx} x^n = n x^{n-1}
$$

### 2. **Sum Rule**

If \( f(x) = u(x) + v(x) \), then:

$$
\frac{d}{dx}[u + v] = \frac{du}{dx} + \frac{dv}{dx}
$$

### 3. **Product Rule**

If \( f(x) = u(x)v(x) \), then:

$$
\frac{d}{dx}[uv] = u \frac{dv}{dx} + v \frac{du}{dx}
$$

### 4. **Quotient Rule**

If \( f(x) = \frac{u(x)}{v(x)} \), then:

$$
\frac{d}{dx}\left[\frac{u}{v}\right] = \frac{v \frac{du}{dx} - u \frac{dv}{dx}}{v^2}
$$

### 5. **Chain Rule**

If \( f(x) = g(h(x)) \), then:

$$
\frac{df}{dx} = \frac{dg}{dh} \cdot \frac{dh}{dx}
$$

Useful for composite functions, e.g., in neural networks.

---

## 📊 Use in Machine Learning

| Area                         | Role of Differentiation                            |
|-----------------------------|-----------------------------------------------------|
| Loss minimization           | Compute gradient to minimize error                 |
| Gradient Descent            | Uses derivative to update weights                  |
| Neural Networks             | Backpropagation uses chain rule                    |
| Convex Optimization         | 2nd derivative indicates global minima             |
| Regularization              | Derivatives involved in adjusting penalty terms    |

---

## 🔁 Practical Example: Gradient Descent

If loss \( L = (y - \hat{y})^2 \), and \( \hat{y} = wx + b \), then:

- Compute \( \frac{dL}{dw} \) and \( \frac{dL}{db} \)
- Use:

  $$
  w := w - \eta \frac{dL}{dw}, \quad b := b - \eta \frac{dL}{db}
  $$

Where \( \eta \) is the learning rate.

---

## 📚 Summary

- Differentiation helps us understand and control **how fast** and in **which direction** we should change model parameters.
- It powers **almost all modern ML algorithms** that rely on optimization.
- Understanding rules of derivatives equips us to handle complex models and custom loss functions.
