# Efficiency in Python

### 1. **What is Efficiency in Python?**
Efficiency in Python refers to how effectively a program uses computational resources, mainly **time** (how fast it executes) and **space** (how much memory it consumes). Writing efficient code ensures the program can handle large inputs and scale well without causing delays or excessive resource usage.

### 2. **Why is Efficiency Important?**
Efficiency is crucial for:
- **Performance**: Efficient code runs faster, which is essential for time-sensitive applications like real-time systems or interactive applications.
- **Resource Management**: It ensures minimal resource usage, which is critical in environments with limited memory or computational power.
- **Scalability**: Efficient code can handle larger datasets or increased user load without significant performance degradation.
- **Cost Reduction**: In cloud computing, reducing memory and CPU usage can lead to lower operational costs.

### 3. **Types of Efficiency**
There are two main types of efficiency to consider in programming:
- **Time Efficiency**: Focuses on how fast the algorithm runs as the input size grows.
- **Space Efficiency**: Focuses on how much memory the algorithm uses as the input size grows.

### 4. **Space and Time Efficiency**
- **Time Efficiency**: Optimizing the algorithm to minimize runtime, which is often expressed in **Big-O notation** (e.g., O(n), O(n^2)).
- **Space Efficiency**: Reducing the amount of memory used by the program, typically considered in **Big-O notation** as well (e.g., O(1) for constant space, O(n) for linear space).

### 5. **Our Focus – Time Efficiency**
Since time efficiency is usually the more pressing concern for many applications (especially when dealing with large datasets), the focus here will be on how to optimize the execution time of an algorithm or program.

### 6. **Techniques to Measure Time**
- **Using the `time` module**: Python’s built-in `time` module provides a simple way to measure execution time:

```python
import time
start_time = time.time()
# Your code here
end_time = time.time()
print(f"Execution time: {end_time - start_time} seconds")
```

- Using `timeit` module: The `timeit` module allows you to measure the time taken by small code snippets with more precision by running them multiple times:

```python
import timeit
execution_time = timeit.timeit("sum(range(1000))", number=1000)
print(f"Execution time: {execution_time} seconds")
```

# Time Complexity Calculation for `fact_iter`

```python
def fact_iter(n):
    """Assumes n is an int >= 0. Computes factorial iteratively."""
    answer = 1
    while n > 1:
        answer *= n 
        n -= 1  
    return answer  
```

### Step 1: Initialization
- `answer = 1` is executed once before entering the loop.
- **Number of operations**: 1.

### Step 2: Loop Condition
- The `while n > 1` condition is checked \( n \) times 
  (once for each iteration, including the last check when \( n = 1 \)).
- **Number of operations**: \( n \).

### Step 3: Inside the Loop
- For each iteration of the loop:
  - `answer *= n`: One multiplication operation.
  - `n -= 1`: One subtraction operation.
- The loop runs \( n-1 \) times because it starts when \( n > 1 \).
- **Number of operations per iteration**: 2.
- **Total number of operations inside the loop**: \( 2(n-1) \).

### Step 4: Return Statement
- `return answer` is executed once after the loop ends.
- **Number of operations**: 1.

---

### Total Operations:
Adding up all operations:
\[
\text{Total Operations} = 1 + n + 2(n-1) + 1
\]

Simplify:
\[
\{Total Operations} = 1 + n + 2n - 2 + 1 = 2n
\]

---

### Time Complexity:
- The total number of operations is proportional to \( n \).
- Dropping constant factors, the time complexity is:
\[
O(n)
\]


### 7. **Measuring Time**
**Best Practice**: For accurate time measurement, it’s better to run the code multiple times and compute the average, which accounts for variability due to system processes.

### 8. **Problems with That Approach**
- **System Dependency**: Time measurements can vary depending on the system’s load, background processes, and hardware.
- **Short-running Code**: For very fast functions, you may not see significant differences in time, making it hard to compare performance.
- **Microbenchmarks**: Measuring execution time in isolation may not represent real-world performance, as other factors like memory access and I/O operations can affect performance.

### 9. **Counting Operations**
Instead of directly measuring runtime, another approach is to count the number of basic operations (e.g., additions, comparisons, iterations) and relate them to the input size. This provides a rough idea of the algorithm's time complexity.

### 10. **What Do We Want to Achieve?**
1. **Evaluate the Algorithm**: Understand the algorithm's performance in terms of how it scales as the input size grows.
2. **Evaluate Scalability**: See how the algorithm performs when handling larger datasets or more complex problems.
3. **Evaluate in Terms of Input Size**: Measure how the time increases as the input size increases, helping identify performance bottlenecks.

### 11. **Different Inputs Change How the Program Runs**
- **Input Size**: As input size grows, the time taken by the algorithm typically increases. The relationship between input size and execution time is described by **time complexity**.
- **Input Characteristics**: The order of input (e.g., sorted vs unsorted) can influence performance. For example, sorting an already sorted list is faster than sorting a randomly ordered list.

### 12. **Orders of Growth**
- **Exact Steps**: Refers to counting the exact number of operations in an algorithm.
- **Big-O Notation (O())**: Asymptotic analysis that describes the upper bound of the growth rate. It ignores constants and lower-order terms to focus on how the algorithm behaves as the input size grows.


### 13. **How to Calculate Order, Types of Order of Growth, and Where That Is Used?**
- **Order of Growth**: This is the rate at which the time or space required by an algorithm grows with respect to the input size.

- **O(1)**: **Constant Time**; The algorithm runs in the same time regardless of input size. It doesn't matter how large the input is; the time taken for the algorithm will remain constant.

```python
def get_first_element(lst):
    return lst[0]
```

- O(n): Linear time; the algorithm's runtime grows linearly with the input size.

```python
def print_elements(lst):
    for element in lst:
        print(element)
```

- O(log n): Logarithmic time; the runtime grows logarithmically with the input size.

```python
def binary_search(arr, x):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

- O(n^2): Quadratic time; the runtime grows quadratically with the input size.

```python
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(0, len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
```

- O(2^n): Exponential time; the runtime doubles with each additional element.

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

- **O(n!)**: **Factorial Time**; The time grows as the factorial of the input size. This is the slowest-growing class of time complexities and is often associated with problems that require checking all permutations (e.g., traveling salesman problem).

### Where These Are Used?

- **O(1)**: Used in simple operations, like accessing an element in a list by index or retrieving a value from a dictionary.
- **O(n)**: Seen in linear operations like iterating over a list or performing a single pass through data.
- **O(log n)**: Common in divide-and-conquer algorithms, such as binary search and certain tree operations (like searching in a balanced binary tree).
- **O(n^2)**: Found in algorithms that involve nested iterations, such as bubble sort or insertion sort.
- **O(2^n)**: Typically appears in brute force solutions for problems involving recursive subproblems, like the Fibonacci sequence or the traveling salesman problem.
- **O(n!)**: Seen in exhaustive search algorithms like those solving the traveling salesman problem using brute force.

### Types of Order of Growth

- **Constant (O(1))**: The time complexity remains the same regardless of the input size.  
  Example: **Array indexing** — Accessing an element by its index in an array.

- **Linear (O(n))**: The time complexity increases linearly with the input size.  
  Example: **Searching** — Iterating through a list to find an element.

- **Quadratic (O(n^2))**: The time complexity increases quadratically with the input size.  
  Example: **Nested loops** — Algorithms like bubble sort or insertion sort that involve iterating through a list and, within that loop, iterating again over the list.

- **Logarithmic (O(log n))**: The time complexity increases logarithmically with the input size.  
  Example: **Binary search** — Searching for an element in a sorted array by dividing the search space in half at each step.

- **n log n (O(n log n))**: The time complexity grows at a rate between linear and quadratic.  
  Example: **Sorting algorithms** — Merge sort, quicksort, and heapsort are typical examples where time complexity is **O(n log n)**.

- **Exponential (O(2^n))**: The time complexity doubles with each additional input element. This complexity grows very quickly and is often impractical for large inputs.  
  Example: **Brute-force solutions** — Problems like the **Fibonacci sequence** (in its naive recursive form) or the **Traveling Salesman Problem** solved using exhaustive search.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define input size
n = np.arange(1, 101)

O_1 = np.ones_like(n) 
O_n = n 
O_n2 = n**2  
O_logn = np.log(n)
O_nlogn = n * np.log(n)
O_2n = 2**n 

fig, axes = plt.subplots(2, 3, figsize=(18, 10))

axes[0, 0].plot(n, O_1, label="O(1) - Constant", color='blue')
axes[0, 0].set_title("O(1) - Constant")
axes[0, 0].set_xlabel("Input Size (n)")
axes[0, 0].set_ylabel("Time Complexity")
axes[0, 0].grid(True)

axes[0, 1].plot(n, O_n, label="O(n) - Linear", color='green')
axes[0, 1].set_title("O(n) - Linear")
axes[0, 1].set_xlabel("Input Size (n)")
axes[0, 1].set_ylabel("Time Complexity")
axes[0, 1].grid(True)

axes[0, 2].plot(n, O_n2, label="O(n^2) - Quadratic", color='red')
axes[0, 2].set_title("O(n^2) - Quadratic")
axes[0, 2].set_xlabel("Input Size (n)")
axes[0, 2].set_ylabel("Time Complexity")
axes[0, 2].grid(True)

axes[1, 0].plot(n, O_logn, label="O(log n) - Logarithmic", color='orange')
axes[1, 0].set_title("O(log n) - Logarithmic")
axes[1, 0].set_xlabel("Input Size (n)")
axes[1, 0].set_ylabel("Time Complexity")
axes[1, 0].grid(True)

axes[1, 1].plot(n, O_nlogn, label="O(n log n) - n log n", color='purple')
axes[1, 1].set_title("O(n log n) - n log n")
axes[1, 1].set_xlabel("Input Size (n)")
axes[1, 1].set_ylabel("Time Complexity")
axes[1, 1].grid(True)

axes[1, 2].plot(n, O_2n, label="O(2^n) - Exponential", color='brown')
axes[1, 2].set_title("O(2^n) - Exponential")
axes[1, 2].set_xlabel("Input Size (n)")
axes[1, 2].set_ylabel("Time Complexity")
axes[1, 2].grid(True)

plt.tight_layout()
plt.show()
```

# 14. Law of Addition
When analyzing the time complexity of two consecutive operations:

If you have two operations, one taking O(n) and another O(m) time, the total time complexity will be O(n + m). If `n` and `m` are of the same order, the time complexity simplifies to O(n).

```python
def add_and_multiply(lst):
    sum_result = sum(lst)  # O(n)
    product_result = 1
    for i in lst:  # O(n)
        product_result *= i
    return sum_result, product_result
```

# 15. Law of Multiplication
When analyzing nested operations:

If you have an outer loop running n times and an inner loop running m times for each iteration of the outer loop, the total time complexity is O(n * m).

```python
def multiply_elements(lst1, lst2):
    result = []
    for i in lst1:  # O(n)
        for j in lst2:  # O(m)
            result.append(i * j)
    return result
```
